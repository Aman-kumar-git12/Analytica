# 🏦 Bank Loan Risk Analysis
### *Theme: Emerald Green & Professional Dark*

Comprehensive exploration of credit metrics, financial correlations, and applicant demographics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Bank Loan Branding Colors (Emerald)
BANK_GREEN = '#00c853'
PURE_BLACK = '#0d1117'
LIGHT_GREEN = '#69f0ae'

# Setting the 'Emerald' Premium Theme
plt.style.use('dark_background')
plt.rcParams.update({
    'axes.facecolor': PURE_BLACK,
    'figure.facecolor': PURE_BLACK,
    'grid.color': '#222222',
    'text.color': '#c9d1d9',
    'axes.labelcolor': '#c9d1d9',
    'xtick.color': '#c9d1d9',
    'ytick.color': '#c9d1d9',
    'font.size': 12
})

df = pd.read_csv('../datasets/bank_loan.csv')
print("Financial Analysis ready.")
```

## 1. Credit Score Overview
Histogram of applicant creditworthiness.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Credit Score'].dropna(), bins=50, color=BANK_GREEN, kde=True)
plt.title('Credit Score Distribution', fontsize=18, color=BANK_GREEN)
plt.show()
```

## 2. Term Distribution
Comparing Short Term vs. Long Term loans.

```python
plt.figure(figsize=(10, 7))
df['Term'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#1b5e20', BANK_GREEN], explode=(0.05, 0), shadow=True)
plt.title('Loan Term Split', fontsize=18, color=BANK_GREEN)
plt.ylabel('')
plt.show()
```

## 3. Annual Income vs. Loan Amount
Correlation between earning capacity and credit request.

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df.sample(1000), x='Annual Income', y='Current Loan Amount', color=BANK_GREEN, alpha=0.5)
plt.title('Income vs Loan Amount', fontsize=18, color=BANK_GREEN)
plt.show()
```

## 4. Housing Status Analysis
How home ownership impacts credit scores.

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Home Ownership', y='Credit Score', color=BANK_GREEN, ci=None)
plt.title('Average Credit Score by Housing', fontsize=18, color=BANK_GREEN)
plt.show()
```

## 5. Feature Correlation Heatmap
Identifying key drivers of financial patterns.

```python
plt.figure(figsize=(12, 10))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='Greens', fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Financial Feature Correlation', fontsize=18, color=BANK_GREEN, pad=20)
plt.show()
```
